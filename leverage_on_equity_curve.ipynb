```python
import pandas as pd, numpy as np
import plotly as plt
import plotly.express as px
import random
```

```python
equity = [369.4,-149.07,293.93,-95.17,-78.51,703.61,768.3,551.7,415.46,501.71,506.61,450.75,444.87,537.98,431.14,351.76,800.64,904.53,-162.8,1138.78,427.22,427.22,351.76,832.99,-80.47,440.95,726.16,671.27,-103.99,-137.31,1761.14,671.27,755.56,608.54,455.65,-116.73,-187.3,-188.28,72.43,62.63]
```

```python
int(sum(equity))
```

```python
equity_l2 = [2*x for x in equity]
equity_l3 = [3*x for x in equity]
equity_l5 = [5*x for x in equity]
equity_l7 = [7*x for x in equity]
```

```python
equity.insert(0, 1000)
equity_l2.insert(0, 1000)
equity_l3.insert(0, 1000)
equity_l5.insert(0, 1000)
equity_l7.insert(0, 1000)
```

```python
d_equity_lev = {
    "l1" : equity,
    "l2" : equity_l2,
    "l3" : equity_l3,
    "l5" : equity_l5,
    "l7" : equity_l7
}
```

```python
df = pd.DataFrame.from_dict(d_equity_lev)
df.reset_index(inplace=True)
df
```

```python
pd.options.plotting.backend = "plotly"
```

```python
# Courbe obtenue pour 12 trades, de 2006 à 2023
df.plot(x=df.index.to_list(), y=[df['l1'].cumsum(), df['l2'].cumsum(), df['l3'].cumsum(), df['l5'].cumsum(), df['l7'].cumsum()])
```

```python
l_res = []

i=0

for i in range (0,50):
    
    temp_list = []
    
    # On démarre après la base 1000. On l'ajoute de nouveau après Shuffle
    temp_list = random.sample(equity_l7[1:], len(equity_l7[1:]))
    temp_list.insert(0, 1000)

    l_res.append(temp_list)
    del temp_list
    
    i+=1
```

```python
# Les 50 aléatoires sont passés en colonnes
df_shuffle = pd.DataFrame(l_res)
df_shuffle = df_shuffle.T
```

```python
# attribue des noms plutôt que les chiffres 0-50 aux noms de colonnes
col_names = ["rand"+str(i) for i in range (0,50)]
df_shuffle.columns = col_names
df_shuffle.tail(1)
```

### Data Viz, Monte Carlo

```python
# Create line plot
fig = px.line(df, x=df_shuffle.index.to_list(), y=df_shuffle['rand0'].cumsum())
```

```python
for i in range (1,50):
    fig.add_scatter(x=df_shuffle.index.to_list(), y=df_shuffle['rand'+str(i)].cumsum())
```

```python
# Display the plot
fig.show()
```
